```python
key = 'sph_fid'
env = MyEnv(key)
front = 'RS'
data = open_rundata(key, 4)
data_fitted = replace_withfit(data)
contribs = data_fitted.drop_duplicates(subset='i')
data_extr = extrapolate_early(data, env)
```

```python
def plot_earlydata(name, normName='', imax=700):
  fig, ax = plt.subplots()
  index = data.index
  val_ini = get_variable(data, name, env)
  val_fit = get_variable(data_fitted, name, env)
  val_extr = get_variable(data_extr, name, env)
  if normName:
    norm = getattr(env, normName)
    (val_ini, val_fit, val_extr) = [val/norm for val in [val_ini, val_fit, val_extr]]
  #cont_ind, val_cont = contribs.index, contribs[name]
  ax.plot(index[index<=imax], val_fit[index<=imax], label='fitted', ls='--')
  ax.plot(index[index<=imax], val_extr[index<=imax], label='extraplolated', ls=':')
  #ax.plot(cont_ind[cont_ind<=imax], val_cont[cont_ind<=imax], label='contributions')
  ylims = ax.get_ylim()
  ax.set_ylim(ylims)
  ax.plot(index[index<=imax], val_ini[index<=imax], label='initial data')
  ax.legend()
```

```python
plot_earlydata('rho')
plt.axhline(env.rho3/env.rhoscale)
```

```python
def plot_withau(name, norm, au_arr=np.logspace(-1, 2, 150)+1, key='sph_fid', logy=True):
  env = MyEnv(key)
  val = np.array([fetch_value_updated(name, 'u4', au*env.u1, env) for au in au_arr])
  valnorm = getattr(env, norm) if norm else 1.
  val *= valnorm
  plt.semilogx(au, val)
  if logy: plt.yscale('log')
```

```python
%%skip
au_arr = np.logspace(-1, 3, 150)
au_arr += 1
env = MyEnv(key)
tcr = np.array([fetch_value_updated('tRS', 'u4', au*env.u1, env) for au in au_arr])
tcr /= env.t0
s = logslope_arr(au_arr, tcr)
plt.loglog(au_arr-1, tcr)
plt.xlabel('$a_u-1$')
plt.ylabel('$t_{cr}/t_0$')
```
