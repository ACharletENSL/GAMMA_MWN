```python
def compare_interpvdata(varname, vartype, front):
  '''
  varname: lfac, ShSt, nu, L
  vartype: inf, alp, s
  '''
  file = './extracted_data/fullsweep_au_' + front + '.csv'
  df = pd.read_csv(file, delimiter='\t')
  df_arr = df.to_numpy().transpose()
  names = ['lfac', 'ShSt', 'nu', 'L']
  types = ['A', 'alpha', 's']
  i_n = names.index(varname)
  i_t = types.index(vartype)
  i = 3*i_n + i_t
  name = varname + '_' + vartype

  logau_data = df_arr[0]
  val_data = df_arr[i]
  logau_arr = np.linspace(logau_data.min(), logau_data.max())
  au_arr = 1 + 10**logau_arr
  fits = [fits_from_au(au, front)[i_n][i_t] for au in au_arr]

  plt.figure()
  plt.scatter(logau_data, val_data, marker='x', label='data')
  plt.plot(logau_arr, fits, label='interp')
  plt.legend()
  plt.xlabel('log$_{10}(a_u-1)$')
  plt.title(name)


```
